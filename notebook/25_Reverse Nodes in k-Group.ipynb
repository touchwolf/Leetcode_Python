# 主要方法
>+ 使用递归的方法来做
>+ 我们用head记录每段的开始位置，cur记录结束位置的下一个节点，然后我们调用reverse函数来将这段翻转，然后得到一个new_head，原来的head就变成了末尾，这时候后面接上递归调用下一段得到的新节点，返回new_head即可

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def reverseKGroup(self, head, k):
        """
        :type head: ListNode
        :type k: int
        :rtype: ListNode
        """
        cur = head
        for _ in range(k):    
            if (not cur):
                return head
            cur = cur.next
        new_head = Solution.reverse(head,cur)#反转之后之前的head就变为了末尾tail
        head.next = Solution.reverseKGroup(self,cur, k)#开始下一段k的反转
        return new_head
    def reverse(head, tail):
        pre = tail#or None
        while(head != tail):
            t = head.next
            head.next = pre
            pre = head
            head = t
        return pre
    #68ms
```
